3) Find polynomial of degree 3 that pass through these four points. Plot to verify.

(1,2) (2,0) (3,2) (7,1)

```python
def w0(x,xs,ys):
    return ys[0]*(((x-xs[1])*(x-xs[2])*(x-xs[3]))/((xs[0]-xs[1])*(xs[0]-xs[2])*(xs[0]-xs[3])))

def w1(x,xs,ys):
    return ys[1]*(((x-xs[0])*(x-xs[2])*(x-xs[3]))/((xs[1]-xs[0])*(xs[1]-xs[2])*(xs[1]-xs[3])))

def w2(x,xs,ys):
    return ys[2]*(((x-xs[0])*(x-xs[1])*(x-xs[3]))/((xs[2]-xs[0])*(xs[2]-xs[1])*(xs[2]-xs[3])))

def w3(x,xs,ys):
    return ys[3]*(((x-xs[0])*(x-xs[1])*(x-xs[2]))/((xs[3]-xs[0])*(xs[3]-xs[1])*(xs[3]-xs[2])))

def f(x,xs,ys):
    return w0(x,xs,ys)+w1(x,xs,ys)+w2(x,xs,ys)+w3(x,xs,ys)

x = np.linspace(0.0,8.0,100)
xs = [1,2,3,7]
ys = [2,0,2,1]
plt.plot(xs,ys,'or')
plt.plot(x,f(x,xs,ys))
plt.axhline(0, color='red')
plt.grid()
```

0) What is $$ \int \limits_{x=1}^{x=3} x^3 \; dx $$

```python
def fi(x):
    return (x**4)/4

def ai(a,b):
    return fi(a)-fi(b)

print ai(3.0,1.0)
```

1)Use Trapezoid Rule to calculate

$$ \int \limits_{x=1}^{x=3} x^3 \; dx $$

Using 1 piece.

```python
def f(x):
    return x**3

def trap(a,b,p):
    h = (b-a)/p
    sum = (f(a)+f(b))/2
    for i in range(1,p):
        sum += f(a+(i*h))
    return sum*h

print trap(1.0,3.0,1)
```

5) Compute the following integral
$$ \int \limits_{x=0}^{x=1} e^{-x} \sin(x^2) \; dx $$

using 100 sub divisions. This is one of those thing you can't do by hands. 

```python
def f(x):
    return math.exp(-x)*math.sin(x**2)

def trap(a,b,p):
    h = (b-a)/p
    sum = (f(a)+f(b))/2
    for i in range(1,p):
        sum += f(a+(i*h))
    return sum*h

print trap(0.0,1.0,100)
```
