0) Finish the implementation of false positon method. Use it to find the solution for $\sin(x) + 1 = x$ do it 20 times.

```python
def f(x):
    return math.sin(x)+1-x

def line(l,r,i):
    m = (f(r)-f(l))/(r-l)
    c = f(r)-(m*r)
    n = -c/m
    
    if (i>20):
        return n
    else:
        if (f(n)*f(l))<0:
            return line(l,n,i+1)
        else:
            return line(n,r,i+1)

print line(1.5,2.5,1)

x = np.linspace(1.5,2.5,100)
y = np.sin(x)-x+1
plt.plot(x,y)
plt.axhline(0, color='black')
plt.axvline(line(1.5,2.5,1), color='orange')
plt.grid()
```

```python
def f(i):
    return math.exp(-i)-i

def func(l,r,i):
    b = (r-l)/2
    g = (r+l)/2
    if (b<=(10**-6)):
        return g
    else:
        if (f(g)*f(l))<0:
            return func(l,g,i+1)
        else:
            return func(g,r,i+1)

print func(0.0,1.0,1)

print "We cannot pick bound A (-1,0), because the solution doesn't exist in that bound."

x = np.linspace(-1,1,100)
y = np.exp(-x)-x
plt.plot(x,y)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.axvline(func(0.0,1.0,1), color='orange')
plt.grid()
```

2) Find all value of $x \in [0,1]$ such that $f(x)=0$ for

$f(x) = -0.000216747 + 0.0072868 x - 0.0977383 x^2 + 0.664685 x^3 - 
 2.40773 x^4 + 4.46766 x^5 - 3.7091 x^6 + x^7$
 
Make sure your answers are at most $10^{-6}$ away from the real solution.

```python
def f(x):
    return -0.000216747 + 0.0072868*x - 0.0977383*x**2 + 0.664685*x**3 - \
            2.40773*x**4 + 4.46766*x**5 - 3.7091*x**6 + x**7

def func(l,r,i):
    b = (r-l)/2
    g = (r+l)/2
    if (b<=(10**-6)):
        return g
    else:
        if (f(g)*f(l))<0:
            return func(l,g,i+1)
        else:
            return func(g,r,i+1)
            
def allsoln(l,r,i):
    diff = r-l
    sd = diff/100
    out = []
    while (l<=r):
        if (f(l)*f(l+sd))<0:
            out.append(func(l,l+sd,1))
        l+=sd
    return out
        
print allsoln(0.0,1.0,1)

x = np.linspace(-1,1,100)
y = f(x)
plt.plot(x,y)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
results = allsoln(0.0,1.0,1)
for i in results:
    plt.axvline(i, color='orange')
plt.ylim(-0.000015,0.004)
# plt.ylim(-0.000015,0.000015)
plt.xlim(0.0,1.0)
plt.grid()
```

2.1) Find all the local minima and maxima of the above function. How do you figure out whether it is a local maximum or local minimum

```python
def fd(x):
    return 0.0072868 - 0.195477*x + 1.99406*(x**2) - 9.63092*(x**3) + 22.3383*(x**4) - 22.2546*(x**5) + 7*(x**6)

def func(l,r,i):
    b = (r-l)/2
    g = (r+l)/2
    if (b<=(10**-6)):
        return g
    else:
        if (fd(g)*fd(l))<0:
            return func(l,g,i+1)
        else:
            return func(g,r,i+1)
            
def allsoln(l,r,i):
    diff = r-l
    sd = diff/100
    minimas = []
    maximas = []
    while (l<=r):
        if (fd(l)*fd(l+sd))<0:
            if (fd(l)>fd(l+sd)):
                maximas.append(func(l,l+sd,1))
            else:
                minimas.append(func(l,l+sd,1))
        l+=sd
    return minimas, maximas

x = np.linspace(-1,1,100)
y = fd(x)
plt.plot(x,y)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
minimas, maximas = allsoln(0.0,1.0,1)

plt.axvline(0, color='orange',label='local maximas')
plt.axvline(0, color='green',label='local minimas')
for i in minimas:
    plt.axvline(i, color='green')
for i in maximas:
    plt.axvline(i, color='orange')
plt.ylim(-0.02,0.02)
plt.xlim(0.0,1.0)
plt.legend()
plt.grid()
```

2.2) What is the `global maximum` and `global minimum` of the function above within the range $x\in[0,1]$

4) Full Width at Half Maximum(FWHM). Distribution width is a very important quantity. It tells you how wide a function is. One way to do that is to use a quantity called full-width half max.

This is the width of the distribution at half maximum. 

For example for distribution shown below the maximum is at x=0.0 and y=1.0

The maximum of this distribution is 1.0. So we are interested in the width of the distribution when the value of the distribution is 1.0/2=0.5

The width of the distribution is shown in black line. In this case it's about 2.35

```python
import math
def demo(x):
    return math.exp(-x**2/2.0)
x = np.linspace(-3,3)
y = [demo(xx) for xx in x]
plt.plot(x,y)
plt.axhline(0.5, color='red')

plt.axvline(2.355/2,color='red',linestyle='dashed')
plt.axvline(-2.355/2,color='red',linestyle='dashed')

plt.grid()
plt.annotate(
    '', xy=(-2.355/2, 0.4), xycoords='data',
    xytext=(2.355/2, 0.4), textcoords='data',
    arrowprops={'arrowstyle': '<->'})
plt.text(0,0.3,'FWHM',horizontalalignment ='center')
```

4.1)Find FWHM of the following function. Make sure the bound on error of the FWHM is less than $10^{-4}$.

```python
import math
def f(x, gamma=3.1):
    mu = 2.345
    return gamma/((x-mu)**2+gamma**2)

def fd(x, gamma=3.1):
    mu = 2.345
    return (2*gamma*(mu-x))/(((x-mu)**2+gamma**2)**2)


def fh(x,n):
    return f(x)-n

def func(l,r,i):
    b = (r-l)/2
    g = (r+l)/2
    if (b<=(10**-4)):
        return g
    else:
        if (fd(g)*fd(l))<0:
            return func(l,g,i+1)
        else:
            return func(g,r,i+1)

def point(l,r,n):
    b = (r-l)/2
    g = (r+l)/2
    if (b<=(10**-4)):
        return g
    else:
        if (fh(g,n)*fh(l,n))<0:
            return point(l,g,n)
        else:
            return point(g,r,n)

plt.axhline(0, color='black')
plt.axvline(0, color='black')

maxima = func(-10.0,10.0,1)
print 'maxima = ' + str(maxima)
half = f(maxima)/2
plt.axhline(half, color='red')

rhm = point(-10.0,10.0,half)
lhm = maxima+(maxima-rhm)
print 'RHM = ' + str(rhm)
print 'LHM = ' + str(lhm)
print 'FWHM = ' + str((maxima-rhm)*2)

x = np.linspace(-10,10,200)
y = [f(xx) for xx in x]
y2 = [fd(xx) for xx in x]
# y3 = [fh(xx,half) for xx in x]
plt.axvline(rhm,color='red',linestyle='dashed')
plt.axvline(lhm,color='red',linestyle='dashed')
plt.axvline(maxima, color='orange',label='maxima')
plt.plot(x,y)
plt.plot(x,y2)
# plt.plot(x,y3)
plt.legend()
plt.annotate(
    '', xy=(lhm, 0.13), xycoords='data',
    xytext=(rhm, 0.13), textcoords='data',
    arrowprops={'arrowstyle': '<->'})
plt.text(maxima,0.1,'FWHM',horizontalalignment ='center')
```

4.2) Plot FWHM(y-axis) as a function of $\gamma$(gamma) from $\gamma=0.5\ldots 3$. Make sure you have at least 30 points.

```python
import math
def f(x, gamma):
    mu = 2.345
    return gamma/((x-mu)**2+gamma**2)

def fd(x, gamma):
    mu = 2.345
    return (2*gamma*(mu-x))/(((x-mu)**2+gamma**2)**2)


def fh(x,n,gamma):
    return f(x,gamma)-n

def func(l,r,i,gamma):
    b = (r-l)/2
    g = (r+l)/2
    if (b<=(10**-4)):
        return g
    else:
        if (fd(g,gamma)*fd(l,gamma))<0:
            return func(l,g,i+1,gamma)
        else:
            return func(g,r,i+1,gamma)

def point(l,r,n,gamma):
    b = (r-l)/2
    g = (r+l)/2
    if (b<=(10**-4)):
        return g
    else:
        if (fh(g,n,gamma)*fh(l,n,gamma))<0:
            return point(l,g,n,gamma)
        else:
            return point(g,r,n,gamma)

def allsoln(l,r,n,gamma):
    diff = r-l
    sd = diff/1000
    out = []
    while (l<=r):
        if (fh(l,n,gamma)*fh(l+sd,n,gamma))<0:
            out.append(func(l,l+sd,n,gamma))
        l+=sd
    return out
        
def fwhm(gamma):
    maxima = func(-10.0,10.0,1,gamma)
    half = f(maxima,gamma)/2
#     print gamma, allsoln(-10.0,10.0,half,gamma)
#     rhm = point(-10.0,10.0,half,gamma)
#     lhm = maxima+(maxima-rhm)
#     return (maxima-rhm)*2
    hm1, hm2 = allsoln(-10.0,10.0,half,gamma)
    return math.fabs(hm1-hm2)
    
    
plt.axhline(0, color='black')
plt.axvline(0, color='black')

# x = np.linspace(0.5,3.0,30) #30 points
# x = np.linspace(0.5,3.0,50) #50 points
# x = np.linspace(0.5,3.0,100) #100 points
x = np.linspace(0.5,3.0,1000) #1000 points
y = [fwhm(xx) for xx in x]
plt.plot(x,y)
plt.grid()
```
