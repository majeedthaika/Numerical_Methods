7.3) Calculate $\pi$ using taylor approximation for 100 **non-zero** terms. (Note I changed this to 100 terms so you must write some sort of for loop to do it).

```python
def pi(terms):
    ans = 1.0
    n = 3
    t = 2
    while (t<=terms):
        if (t%2==0):
            ans -= 1.0/n
        else:
            ans += 1.0/n
        n+=2
        t+=1
    return ans
print "pi approx. to 100 non-zero terms = " + str(pi(100)*4)
```
